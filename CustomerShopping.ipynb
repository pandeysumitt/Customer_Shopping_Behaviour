# Treating the null values

```python
df['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in Electronics Category
df.loc[df['Category'] == 'Electronics', 'Size'] = 'Not Applicable'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in Accessories Category
df.loc[df['Category'] == 'Accessories', 'Size'] = 'Free Size'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in Footwear Category
df.loc[df['Category'] == 'Footwear', 'Size'] = 'Not Available'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in Footwear Category
df.loc[(df['Category'] == 'Clothing') &(df['Size'].isnull()), "Size"] = df[df["Category"] == "Clothing"]["Size"].mode()[0 ]
```

```python
df.groupby('Category')['Size'].unique()
```

# Review Rating

```python
df.head()
```

```python
# mean of review ratings
df['Review Rating'].mean()
```

```python
df.groupby("Item Purchased")['Review Rating'].mean()
```

```python
# filling null values of Review Ratings with the mean of each Item Purchased
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df['Review Rating'] = df['Review Rating'].round(2)
```

```python
df.head(10)
```

# Previous Purchases

```python
# filling previous purchases with Zero(0)
df['Previous Purchases']= df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

# Purchase Amount(USD)

```python
df.groupby('Item Purchased')['Purchase Amount (USD)'].mean()
```

```python
# filling null values of Purchase Amount (USD) with the mean of each Item Purchased
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```

# Category

```python
# Fill remaining null values in 'Category' with the overall mode of the 'Category' column
if df['Category'].isnull().any():
    overall_category_mode = df['Category'].mode()[0]
    df['Category'] = df['Category'].fillna(overall_category_mode)
```

```python
df.isnull().sum()
```

# Renaming the Headers

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()
```
